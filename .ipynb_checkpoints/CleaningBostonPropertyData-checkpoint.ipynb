### 2.3 Parsing Number of Bedrooms and Number of Bathrooms

**Parsing and extracting bedrooms and bathrooms.** To extract a property's numer of bedrooms and bathrooms, I have used regular expressions. The rooms column has two different formats:

- Number of bathrooms:[data];Number of bedrooms:[data]
- Number of bedrooms:[data];Number of bathrooms:[data]

The following regular expressions are used to extract the data:
- *'Number of bathrooms: (\d+.\d+)'* 
<br/> To extract the bathrooms value, I have to capture the decimal/floating point value that is right after the string 'Number of bathrooms'.
- *'Number of bedrooms: (\d+)'* 
<br/> To extract the bedrooms value, I have to capture the whole number that is right after the string 'Number of bedrooms:'.


If there is a problem during the extraction, the value will be set to None, similar to `errors = 'coerce'`.

```python
parseDF.head()
```

```python
def extract_bathrooms(line):
    # stricter : \d.(?:[05]|[27][5])
    # lenient for now, as we're just parsing. we'll correct these values later
    x = re.findall('Number of bathrooms: (\d+.\d+)', line)
    if x :
        return(x[0])
    else:
        return None


def extract_bedrooms(line):
    x = re.findall('Number of bedrooms: (\d+)', line)
    if x :
        return(x[0])
    else:
        return None

parseDF['bedrooms'] = parseDF['rooms'].map(lambda x: extract_bedrooms(x))
parseDF['bathrooms'] = parseDF['rooms'].map(lambda x: extract_bathrooms(x))
```

To verify if what I have extracted is in the numeric data type, I again parsed using the following steps:

1. Parse using Pandas' `to_numeric()` method, with errors parameter set to `coerce`. This will set problematic rows as `NaN`.
2. Check if there are any null values.
3. Correct the null values.

```python
parseDF['bedrooms'] = pd.to_numeric(parseDF['bedrooms'], errors = 'coerce')
parseDF['bathrooms'] = pd.to_numeric(parseDF['bathrooms'], errors = 'coerce')
print parseDF[parseDF['bedrooms'].isnull()]
print parseDF[parseDF['bathrooms'].isnull()]
```

Since there are no null values, all the values in bedrooms and bathrooms are in numeric data type, so no problems in the domain.

### 2.4 Parsing Sqft_Living and Sqft_Lot

**Parsing and extracting sqft_living and sqft_lot.** To extract a property's sqft_living and sqft_lot, I have continued using Python's regular expressions library `re`. The rooms column has a single format:

- 	sqft_living/sqft_lot=[data]\ [data]

The following regular expressions are used to extract the data:
- *'sqft_living/sqft_lot=(\d+)'* 
<br/> To extract the sqft_living value, I have to capture the whole number right after the string 'sqft_living/sqft_lot='.

- *'sqft_living/sqft_lot=(?:\d+)\\\ (\d+)'* 
<br/> To extract the sqft_lot value, I have to capture the whole number that is right after the string 'sqft_living/sqft_lot = [sqft_living]\ '. Note that I specified that the first set of digits must not be included, so there's a non capturing group (?:\d+). I also have to enforce \\\ to denote a single \, while the data I want is in the capturing group (\d+).


If there is a problem during the extraction, the value will be set to None, similar to `errors = 'coerce'`.



```python
def extract_sqft_living(line):
    x = re.findall('sqft_living/sqft_lot=(\d+)', line)
    if x :
        return(x[0])
    else:
        return None


def extract_sqft_lot(line):
    x = re.findall('sqft_living/sqft_lot=(?:\d+)\\\ (\d+)', line)
    if x :
        return(x[0])
    else:
        return None

parseDF['sqft_living'] = parseDF['area.sqft_living/sqft_lot'].map(lambda x: extract_sqft_living(x))
parseDF['sqft_lot'] = parseDF['area.sqft_living/sqft_lot'].map(lambda x: extract_sqft_lot(x))
```

To verify if what I have extracted is in the numeric data type, I just followed what I did in the price column. Since there were no null values upon parsing, both columns' values came from the domain (of whole numbers).

```python
# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_numeric.html#pandas.to_numeric
parseDF['sqft_living'] = pd.to_numeric(parseDF['sqft_living'], errors = 'coerce')
parseDF['sqft_lot'] = pd.to_numeric(parseDF['sqft_lot'], errors = 'coerce')
print parseDF[parseDF['sqft_living'].isnull()]
print parseDF[parseDF['sqft_lot'].isnull()]
```

### 2.10 Parsing Street, City, Statezip, and Country

**Parsing and extracting street, city, statezip and country.** To extract a property's street, city, statezip and country from the address column, I have used regular expressions' `re.findall()`. The address column has the format:

- [street], [city], [statezip], [country]

The following regular expressions are used to extract the data:

- *'([\w][\s\w\-\/]+)'* 

<br/> This regular expression means, extract all the strings that strictly start with an *alphanumeric character* (to avoid the need to strip the values) followed by one to many instances of *alphanumeric character*, *spaces*, *dash (-)*, or *forward slash (/)*. 
<br/> The street corresponds to the first matched group, city the second matched group, statezip the third matched group, and the country the fourth matched group.

If there is a problem during the extraction, the value will be set to None, similar to `errors = 'coerce'`.



```python
# address may contain space, word, -, or /
# but it MUST NOT start with a space
def parse_address(line, part):
    x = re.findall('([\w][\s\w\-\/]+)', line)
    if len(x) != 4:
        return None
    elif x  and part >= 0 and part <= 3:
        return(x[part])
    
# extract the street by passing the whole address and getting the first matched group
parseDF['street'] = parseDF['address'].map(lambda x: parse_address(x, 0))

# extract the city by passing the whole address and getting the second matched group
parseDF['city'] = parseDF['address'].map(lambda x: parse_address(x, 1))

# extract the statezip by passing the whole address and getting the third matched group
parseDF['statezip'] = parseDF['address'].map(lambda x: parse_address(x, 2))

# extract the country by passing the whole address and getting the fourth matched group
parseDF['country'] = parseDF['address'].map(lambda x: parse_address(x, 3))
```

#### Univariate analysis of outliers
For the remaining numerical variables, I will also check the computed outlier bounds based from the IQR method introduced in the lecture. The IQR method is a helpful method in determining the outliers since it does not depend on the mean or the standard deviation.(Purplemath, 2017) The original formula goes like this:

<br/> upperbound = Q3 + 3 \* IQR (upper outer fence - i.e. any value greater than this is an outlier)
<br /> lowerbound = Q1 - 3 \* IQR (lower outer fence - i.e. any value lower than this is an outlier)

However, the main problem with this is that this assumes the dataset's distribution to be normal/symmetric. As shown in the exploratory analysis above, our dataset is highly skewed to the left, so I have adjusted the threshold parameter from 3 to 4, since most of the outliers will be further than both the mean and the median of the sample. I also just get the upperbound outlier since there will be no lowerbound outliers if it's skewed this way.  

```python
# create a function to determine the upper outer fence outlier
def det_outliers(column):
    q1 = column.quantile(.25)
    q3 = column.quantile(.75)
    iqr = q3 - q1
    # t is large to account for the fact that the data is positively skewed
    upperBound = q3 + 4 * iqr
    
    return upperBound
    
```

### Price Prediction

To fit a Linear Regression Model I used sklearn's **LinearRegression** library. After checking out all possible combinations, the features I've decided to add are 'sqft_living', 'sqft_lot', 'bathrooms', 'bedrooms', 'view', 'floors', 'condition', and 'price', as this combination yields the highest r-squared value when the model is scored.

Again, sklearn's Linear Regression is done in two steps:
1. Fitting the data to form a model via the `fit()` function, using the feature vector (i.e. the dataframe with all numerical variables in the dataset)
2. Predicting the target value using the fit via the `predict()` function

To learn more about Linear Regression, refer to <a href = 'http://scikit-learn.org/stable/auto_examples/linear_model/plot_ols.html'>here</a>. (Scikit-Learn Developers, 2017a)

Another good reference for Linear Regression in Python, which I based my code upon, can be found <a href = 'http://www.bigdataexaminer.com/2016/03/31/how-to-run-linear-regression-in-python-scikit-learn/'>here.</a> (Jeevan, 2016)


I will use the model to predict the price if the price is equal to 0.

```python
# use linear regression to fit the data

# drop the rows we're trying to predict
dfModel = df[df['price'] > 0]
# drop all the non-scalar variables and the target variable -- price
dfModel = dfModel[['sqft_living', 'sqft_lot', 'bathrooms', 'bedrooms', 'view', 'floors', 'condition', 'price']]

# use this filtered dataframe during prediction
dfPredictor = df.copy()
dfPredictor = dfPredictor[['sqft_living', 'sqft_lot', 'bathrooms', 'bedrooms', 'view', 'floors', 'condition']]
```

```python
dfModel.shape
```

```python
len(dfModel['price'])
```

```python
# make sure our feature dataframe do not have any 0 prices
dfModel.tail()
```

```python
lm = LinearRegression()

# create a linear regression model from the feature dataframe
lm.fit(dfModel[[x for x in dfModel.columns if x != 'price']], dfModel['price']) #fit

# check the score
print lm.score(dfModel[[x for x in dfModel.columns if x != 'price']], dfModel['price'])
```

What is this score? This score is called the **r-squared**, which measures how close the data are to the fitted regression line. A measure of 0% indicates that the model does not explain the variability of the dataset around the mean, while a measure of 100% indicates that the model completely explains the dataset variability. (Frost, 2013) This r-squared score is okay--not too high but not too low either. This may be because of the general variability of the dataset.

```python
dfCopy = df.copy()

dfCopy.loc[df['price'] == 0, 'price'] = lm.predict(dfPredictor)

# round off the price since the price has int datatype, and price values have no decimals
df['price'] = dfCopy['price'].round(0)
```

```python
df[['price']].tail(100)
```

Checking the scatterplot, we can see that the dataset with updated prices still follows the linear relationship as expected.

```python
plt.plot(dfCopy['price'], dfCopy['sqft_living'], "o")
plt.xlabel('price')
plt.ylabel('sqft_living')
```

The final data frame is now ready. For the final touches, let's just bring back the look and feel of the data to how it came from. (Python Software Foundation, 2017d) 

```python
# reorder the columns for the final csv
dfOutput = df[['date', 'price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'street', 'city', 'statezip', 'country']]

# convert the yr_built into string %Y format
dfOutput['yr_built'] = [d.strftime('%Y') if not pd.isnull(d) else '' for d in dfOutput['yr_built']]

# convert the yr_renovated into string %Y, otherwise if there's no value set as empty
dfOutput['yr_renovated'] = [d.strftime('%Y') if not pd.isnull(d) else '' for d in dfOutput['yr_renovated']]

# convert the date into string %Y%m%dT%H%M%S format
dfOutput['date'] = [d.strftime('%Y%m%dT%H%M%S') if not pd.isnull(d) else '' for d in dfOutput['date']]
```
